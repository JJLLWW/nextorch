```python
from nextorch import bo, io, plotting
```

```python
X_real = np.array([[1.0, 1.0, 1.0], [5.0, 5.0, 5.0], [25.0, 25.0, 25.0]])
Y_real = np.array([[40.670, 41.717, 6.897]]).T
X_names = ["L_stub1", "L_stub2", "L_stub3"]
X_units = ["mm", "mm", "mm"]
X_ranges = [[1.0, 25.0], [1.0, 25.0], [1.0, 25.0]]
```

```python
exp_comsol = bo.COMSOLExperiment("Stub tuner")
```

```python
file_name = "../Stub_tuner/comsol_example"
comsol_location =  "/home/tychen/comsol54/multiphysics/bin/comsol"
output_file = "../Stub_tuner/simulation_result.csv"

exp_comsol.input_data(X_real, Y_real, X_names=X_names, X_units=X_units, X_ranges=X_ranges, decimals=2)
exp_comsol.set_optim_specs(file_name, comsol_location, output_file, 2)
exp_comsol.run_trials_auto(n_trials=20)
```

```python
x1_fixed_real = 15
x2_fixed_real = 15
x3_fixed_real = 15

print('Reponses at fixed x1, vary x2 and x3: ')
plotting.response_heatmap_exp(exp_comsol, x_indices = [1, 2], 
                              fixed_values_real = x1_fixed_real)
print('Reponses at fixed x2, vary x1 and x3: ')
plotting.response_heatmap_exp(exp_comsol, x_indices = [0, 2],
                              fixed_values_real = x2_fixed_real)
print('Reponses at fixed x3, vary x1 and x2: ')
plotting.response_heatmap_exp(exp_comsol, x_indices = [0, 1],
                              fixed_values_real = x3_fixed_real)
```

```python
plotting.opt_per_trial_exp(exp_comsol)
```
